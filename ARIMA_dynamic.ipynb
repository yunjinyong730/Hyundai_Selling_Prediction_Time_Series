```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
import itertools

# Load the data
file_path = 'month_updated_transposed_cleaned.xlsx'
data = pd.read_excel(file_path)

# Standardize column names to avoid KeyErrors
data.columns = data.columns.str.strip()

# Set 'Date' as the index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Forecast sales for each model using ARIMA
forecast_periods = (2030 - 2024) * 12  # Forecast until 2030, starting from the end of 2024

# Function to tune ARIMA hyperparameters manually and fit the model
def tune_arima(series, p_values, d_values, q_values):
    best_score, best_cfg = float("inf"), None
    for p, d, q in itertools.product(p_values, d_values, q_values):
        try:
            order = (p, d, q)
            model = ARIMA(series, order=order)
            model_fit = model.fit()
            aic = model_fit.aic
            if aic < best_score:
                best_score, best_cfg = aic, order
        except:
            continue
    return best_cfg

# Function to fit ARIMA model and make future predictions
def forecast_arima(series, model_order, steps):
    model = ARIMA(series, order=model_order)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    return forecast, model_fit

# Forecast and plot each model
plt.figure(figsize=(15, 10))

with open('sales_forecast.txt', 'w') as file:
    for i, model in enumerate(['Avante', 'Sonata', 'G80']):
        if model in data.columns:
            plt.subplot(3, 1, i+1)
            series = data[model]

            # Train/test split to evaluate model accuracy
            split_point = int(len(series) * 0.8)
            train, test = series[:split_point], series[split_point:]

            # Tune ARIMA hyperparameters
            p_values = range(0, 4)
            d_values = range(0, 3)
            q_values = range(0, 4)
            best_order = tune_arima(train, p_values, d_values, q_values)

            # Forecast using training data
            forecast, model_fit = forecast_arima(train, best_order, len(test))

            # Calculate and display RMSE
            rmse = np.sqrt(mean_squared_error(test, forecast))
            print(f'RMSE for {model} with order {best_order}: {rmse:.2f}')
            file.write(f'RMSE for {model} with order {best_order}: {rmse:.2f}\n')

            # Refit on the entire series for future forecast
            full_forecast, _ = forecast_arima(series, best_order, forecast_periods)

            # Create future dates for the forecast
            last_date = series.index[-1]
            future_dates = pd.date_range(last_date + pd.DateOffset(months=1), periods=forecast_periods, freq='M')

            # Print and save forecast values
            forecast_series = pd.Series(full_forecast, index=future_dates)
            print(f'Forecast for {model} from 2024/10 to 2030:')
            print(forecast_series['2024-10-01':])
            file.write(f'Forecast for {model} from 2024/10 to 2030:\n')
            file.write(forecast_series['2024-10-01':].to_string())
            file.write('\n\n')

            # Plot historical data
            plt.plot(series, label='Historical')

            # Plot forecast data
            plt.plot(future_dates, full_forecast, label='Forecast', color='red')
            plt.title(f'{model} Sales Forecast (RMSE: {rmse:.2f})')
            plt.legend()
            plt.xlabel('Date')
            plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
